```python

logo = """
                            ___.                                                      .__                
  ____  __ __  _____\_ |__   ___________     ____  __ __   ____   ______ _____|__| ____    ____  
 /    \|  |  \/     \| __ \_/ __ \_  __ \   / ___\|  |  \_/ __ \ /  ___//  ___/  |/    \  / ___\ 
|   |  \  |  /  Y Y  \ \_\ \  ___/|  | \/  / /_/  >  |  /\  ___/ \___ \ \___ \|  |   |  \/ /_/  >
|___|  /____/|__|_|  /___  /\___  >__|     \___  /|____/  \___  >____  >____  >__|___|  /\___  / 
     \/            \/    \/     \/        /_____/             \/     \/     \/        \//_____/  
                                                                                                 
                                                                                                 
                                                                                                
"""

import random
from IPython.display import clear_output
```

```python
print(logo)
def random_number():
    return random.randint(1,100)

def life(no_of_lives):
    if no_of_lives == 'hard':
        return 5
    else:
        return 10



def score(number,guess):
    if guess > number:
        return 1
    elif guess < number:
        return -1
    else:
        return 0

play_again = 'yes'
while play_again == 'yes':
    number = random_number()
    lives  = life(input("Enter the level hard or easy you wanna play?" ).lower())

    while lives > 0:
        guess = int(input("Enter the no you want to guess? "))
        your_score = score(number,guess)
        if your_score == 1:
            print("number too high")
            lives-=1
        elif your_score == -1:
            print("number too low")
            lives-=1
        
        else:
            print("Correct guess")
            play_again = input("Do you want to play again yes no? ").lower()
            if play_again == 'no':
                break
            else:
                clear_output(wait=True)
                play_again == 'yes'
                
    if lives == 0:
            print(f"Your guess chance are 0 number is {number}")
            break
    

while input("Do you want to play again? ") == 'yes':
    clear_output(wait = True)
    play_again = 'yes'
    
print("Have a nice day!!!!!!!")
```
